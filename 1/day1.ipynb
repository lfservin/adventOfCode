# Parse string front and backwards to retrieve the first digit in every direction


```python
import re


file = open("input.txt")

list = []

for line in file:
    #do something
    dstart = re.search(r"\d", line)
    dend = re.search('.*([0-9])[^0-9]*$', line)
    print(line)
    value="{}{}".format(dstart.group(0),dend.group(1))
    print(value)
    list.append(int(value))
```
